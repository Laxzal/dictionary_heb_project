```python
dictionary_df = pd.DataFrame(columns=["Heb","Eng"])
```

```python
for link in soup.findAll(True, {"class":["menukad", "dict-meaning"]}):
    if link.string == 'Meaning':
        continue
    if link.string == "שִׁנִּיתָ":
        continue
    if link.string== "שִׁינִּיתָ":
        continue
    if link.string =="שינית":
        continue
    print (link.string)
    
```

```python
for link in soup.findAll(True, {"class":["menukad"]}):
    if link.string == 'Meaning':
        continue
    if link.string == "שִׁנִּיתָ":
        continue
    if link.string== "שִׁינִּיתָ":
        continue
    if link.string =="שינית":
        continue
    print(link.string)
    dictionary_df = dictionary_df.append({"Heb":link.string}, ignore_index=True)
```

```python
for link in soup.findAll(True, {"class":["dict-meaning"]}):
    if link.string == 'Meaning':
        continue
    if link.string == "שִׁנִּיתָ":
        continue
    if link.string== "שִׁינִּיתָ":
        continue
    if link.string =="שינית":
        continue
    print(link.string)
    dictionary_df = dictionary_df.append({"Eng":link.string}, ignore_index=True)
```
