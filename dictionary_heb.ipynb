```python
dictionary_df = pd.DataFrame(columns=["Heb","Eng"])

import time

start = time.time()

for i in list_link:
    print(str(list_link.index(i)) + "/" + str(len(list_link)))

    verb_link_source = requests.get(str(i))
    if verb_link_source.status_code == 200:
        verb_soup = BeautifulSoup(verb_link_source.content, 'html.parser')

        list_id =[]
        verb_id = verb_soup.findAll(True,{'class':['conj-td']})
        for link in verb_id:
            for x in link.findAll("div", id=True):
                list_id.append(str(x['id']))


        for y in list_id:
            for link_word in verb_soup.findAll(True, {"id":[y]}):
                if link_word.findAll(True,{"class":["chaser"]}):
                    continue
                
                else:
                    if link_word.find(True,{"class":['meaning']}) is None:
                        for z in link_word.findAll(True,{"class":['menukad']}):
                            dictionary_df = dictionary_df.append({"Heb":z.text,
                                                    "Eng":str(verb_soup.find(True,{"class":['lead']}).text)}, 
                                                         ignore_index=True)
                    else:
                        for z in link_word.findAll(True,{"class":['menukad']}):
                        #print (z.text)
            #print(link.find(True,{"class":['menukad']}).text, link.find(True,{"class":['menukad']}).next.next.next.next.next)
                        #print()        


                            dictionary_df = dictionary_df.append({"Heb":z.text,
                                                    "Eng":str(link_word.find(True,{"class":['meaning']}).text)}, 
                                                         ignore_index=True)  

end = time.time()

print (end-start)
```
