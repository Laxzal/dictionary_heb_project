```python
for i in range(res[0]+1):
    if i ==0:
        continue
    print(i)
```

```python
dictionary_df = pd.DataFrame(columns=["Heb","Eng"])
for i in range(res[0]+1):
    print(str(i) + "/" + str(res[0]))
    if i==0:
        continue
    source = requests.get('https://www.pealim.com/dict/?page=' + str(i))
    #soup = BeautifulSoup(source, 'lxml')
    soup = BeautifulSoup(source.content, 'html.parser')
    for link,test in zip(soup.findAll(True, {"class":["menukad"]}), soup.findAll(True, {"class":["dict-meaning"]})):
        if link.string == 'Meaning':
            continue
        if link.string == "שִׁנִּיתָ":
            continue
        if link.string== "שִׁינִּיתָ":
            continue
        if link.string =="שינית":
            continue
        if test.string == 'Meaning':
            continue
        if test.string == "שִׁנִּיתָ":
            continue
        if test.string== "שִׁינִּיתָ":
            continue
        if test.string =="שינית":
            continue
        #print (link.string)
        #print (test.string)
        dictionary_df = dictionary_df.append({"Heb":link.string,
                                             "Eng":test.string}, 
                                             ignore_index=True)  
```

```python
def remove_niqqud_from_string(my_string):
    return ''.join(['' if  1456 <= ord(c) <= 1479 else c for c in my_string])
```

```python
string = 'אַבָּא'
```

```python
remove_niqqud_from_string(string)
```
